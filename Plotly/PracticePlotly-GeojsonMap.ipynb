# Importing Packages

> Importing Pandas , Plotly.express and json

```
import plotly.express as px
import pandas as pd
import json
```

# Without merging the geojson file with csv

```
indian_states = json.load(open("/content/drive/My Drive/Colab Notebooks/Files_To_Read/states_india.geojson" , "r"))
indian_states
```

```
population = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Files_To_Read/IndianPoputlaion.csv")
population.head()
```

```
fig = px.choropleth(population , 
                    geojson=indian_states , 
                    locations='State or union territory' ,
                    featureidkey='properties.st_nm' ,
                    projection="mercator" , 
                    color="Area[16]")
fig.update_geos(fitbounds='locations' , visible=False)
fig.show()
```

# Merging geojson file with csv file

> Loading geojson file

```
indian_states2 = json.load(open("/content/drive/My Drive/Colab Notebooks/Files_To_Read/states_india.geojson" , "r"))
indian_states2
```

> Providing "id" same as statecode in the geojson file

```
state_id_map = {}
for feature in indian_states2['features']:
  feature['id'] = feature['properties']['state_code']
  state_id_map[feature['properties']['st_nm']] = feature['id']

indian_states2
```

> Dictionary named "state_id_map" to add the "id" to the Csv

```
state_id_map
```

> Reading Csv file

```
population2 = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Files_To_Read/IndianPoputlaion.csv")
population2.head()
```

> Manipulating Csv file , Creating a new Density column and renaming some State names according to geojson file

```
population2['Density'] = population2['Density[a]'].apply(lambda x : int(x.split("/")[0].split(".")[0].replace(",","")))
# # print(population2.loc[25])
population2.loc[22,'State or union territory'] = population2.loc[22,'State or union territory'].replace("[c]","")
population2.loc[25,'State or union territory'] = "Arunanchal Pradesh"
population2.loc[25,'State or union territory'] = "Andaman & Nicobar Island"
population2.head()
```

> Providing "id" in the Csv file as same as geojson file , so that they can map

```
population2['id'] = population2['State or union territory'].apply(lambda x : state_id_map[x] if state_id_map.get(x)!=None else 0 )
population2.loc[28,'id'] = 7
population2
```

> Plotting the choropleth map with the help of geojson and csv files.

```
fig = px.choropleth(population2 , locations="id" , color="Density" , geojson=indian_states2)
fig.update_geos(fitbounds="locations" , visible=False)
fig.show()
```
