Importing Bokeh

```
from bokeh.models import GeoJSONDataSource,HoverTool, LinearColorMapper
from bokeh.plotting import figure, output_file, show
from bokeh.tile_providers import CARTODBPOSITRON, get_provider
import bokeh.io
bokeh.io.output_notebook() # Displying the ma inside the notebook
import json
import pandas as pd
from bokeh.palettes import brewer
```

```
# output_file("tille.html")
```

```
tile_provider = get_provider(CARTODBPOSITRON)
```

```
tools = 'wheel_zoom,pan,reset'
p = figure(x_range=(-2000000, 4000000), y_range=(-1000000, 7000000),
           x_axis_type="mercator", y_axis_type="mercator" , tools=tools)
p.add_tile(tile_provider)

show(p)
```

```
indian_states = json.load(open("/content/drive/My Drive/Colab Notebooks/Files_To_Read/states_india.geojson" , "r"))
len(indian_states['features'])
```

```
for i in range(len(indian_states['features'])):
  indian_states['features'][i]['properties']['color'] = ['blue','red','green'][i%3]

indian_states['features']
```

```
geo_source = GeoJSONDataSource(geojson=json.dumps(indian_states))
```

```
# cmap = LinearColorMapper(palette=Viridis256)
TOOLS = 'pan,wheel_zoom,box_zoom'
p = figure(title="India Map" ,tools=TOOLS, width=800 , height=800)
p.grid.grid_line_color = None # will not show the grid behind the image. 
p.add_tools(HoverTool(tooltips=[('State','@st_nm'),('StateCode', '@state_code')]))
p.axis.visible = False
p.patches('xs' , 'ys' ,fill_alpha=0.7, line_color='black' , fill_color='color', line_width=0.1, source=geo_source)
show(p)
```
