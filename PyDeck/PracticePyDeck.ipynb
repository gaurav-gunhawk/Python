```
import pydeck as pdk
import pandas as pd
```

```
# UK_ACCIDENTS_DATA = ('https://raw.githubusercontent.com/uber-common/'
#                     'deck.gl-data/master/examples/3d-heatmap/heatmap-data.csv')
```

```
UK_ACCIDENTS_DATA  = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Files_To_Read/indian_states_lat_long.csv")
UK_ACCIDENTS_DATA = UK_ACCIDENTS_DATA[['lat' , 'lng']]
UK_ACCIDENTS_DATA
```

```
layer = pdk.Layer(
    'ScatterplotLayer',
    UK_ACCIDENTS_DATA,
    get_position=['lng', 'lat'],
    auto_highlight=True,
    get_radius=10000,
    get_fill_color='[180, 0, 200, 140]',
    pickable=True)

view_state = pdk.ViewState(
    longitude=78.962883,
    latitude=20.593683,
    zoom=6,
    min_zoom=5,
    max_zoom=15,
    pitch=40.5,
    bearing=-27.36)

# Combined all of it and render a viewport
r = pdk.Deck(layers=[layer], initial_view_state=view_state , mapbox_key="pk.eyJ1IjoiZ2F1cmF2Z3VwdGEwNyIsImEiOiJja2QxdXo5dDkwdWQ4MzZxbXN5a3k1MnFvIn0.qf2qnag9B9uU6OoONVokPQ")
r.to_html('demomo.html', notebook_display=False)
# r.show()
```
